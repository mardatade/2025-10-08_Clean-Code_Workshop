## Global Ocean Biogeochemistry Hindcast

```python
ds_gobh = copernicusmarine.open_dataset(
    dataset_id="cmems_mod_glo_bgc_my_0.25deg_P1M-m"
)
ds_gobh
```

```python
ds_gobh_coarse = ds_gobh.isel(depth=slice(None, None, 15)).sel(time=slice("1995", "1996"))
ds_gobh_coarse
```

```python
%%time

ds_gobh_coarse = ds_gobh_coarse.compute()
```

```python
ds_gobh_coarse.drop_encoding().to_netcdf("data/cmems_mod_glo_bgc_my_0.25deg_P1M-m_depth-skip-15_time-1995-1996.nc")
```

## Global Ocean Ensemble Physics Reanalysis

```python
ds_goepr = copernicusmarine.open_dataset(dataset_id="cmems_mod_glo_phy-mnstd_my_0.25deg_P1M-m")
ds_goepr
```

```python
ds_goepr = ds_goepr.drop([v for v in ds_goepr.data_vars if "_std" in v])
ds_goepr = ds_goepr.rename_vars({v: v[:-5] for v in ds_goepr.data_vars if "_mean" in v})
ds_goepr
```

```python
ds_goepr_coarse = ds_goepr.isel(depth=slice(None, None, 15)).sel(time=slice("1995", "1996"))
ds_goepr_coarse
```

```python
%%time

ds_goepr_coarse = ds_goepr_coarse.compute()
```

```python
ds_goepr_coarse.drop_encoding().to_netcdf("data/cmems_mod_glo_phy-mnstd_my_0.25deg_P1M-m_depth-skip-15_time-1995-1996.nc")
```
